```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ['c++', 'python', 'java', 'c#', 'javascript', 'php', 'ruby', 'r', 'swift', 'go']
y = [20, 15, 10, 7, 5, 4, 3, 2, 1, 1]
print(len(x))
print(len(y))
```

```python
plt.bar(x, y,color='red',align='edge',width=0.6,edgecolor='black',lw=1)
plt.show()
'''
x = x ekseni
y = y ekseni
color = barların rengi
align = barların x eksenindeki konumu
width = barların kalınlığı
edgecolor = barların kenar rengi
lw = barların kenar kalınlığı
'''
```
