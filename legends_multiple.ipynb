```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock_a=np.random.normal(0,100,10)
stock_b=np.random.normal(0,100,10)
stock_c=np.random.normal(0,100,10)
```

```python
plt.plot(stock_a,label='company_a')
plt.plot(stock_b,label='company_b')
plt.plot(stock_c,label='company_c')
plt.legend(loc='lower left')
'''
plt.legend = anlamı grafiğin hangi çizgiye 
ait olduğunu gösteren bir tablo oluşturur.
'''
```

```python
votes=[10,2,5,6,7,8,9,10]
people=['a','b','c','d','e','f','g','h']
```

```python
plt.pie(votes,labels=None)
plt.legend(labels=people,loc='lower left')
plt.show()
```
