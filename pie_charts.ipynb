```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
languages = ['Python', 'SQL', 'Java', 'Go', 'JavaScript', 'C++', 'Dart', 'R',
                'Swift', 'PHP']
votes = [26, 19, 8, 7, 7, 6, 5, 5, 4, 2]
print(len(languages), len(votes))
explode = [0.0, 0, 0, 0, 0.3, 0, 0, 0, 0, 0]
```

```python
plt.pie(votes, labels=languages,explode=explode,
        autopct="%.1f%%",
        pctdistance=0.5,
        shadow=False,
        startangle=130)
plt.show()
'''
labels = dilimlerin isimleri
autopct = yüzde değerlerini gösteriyor
pctdistance = yüzde değerlerinin uzaklığı
shadow = gölge
startangle = başlangıç açısı
explode = dilimlerin arasındaki boşluklar
'''
```
