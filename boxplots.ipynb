```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
heights = np.random.normal(172,8,300)
print(heights)
```

```python
plt.boxplot(heights)
plt.show()
```

```python
first_quartile = np.percentile(heights,25)
second_quartile = np.percentile(heights,50)
third_quartile = np.percentile(heights,75)
fourth_quartile = np.percentile(heights,100)
print(first_quartile)
print(second_quartile)
print(third_quartile)
print(fourth_quartile)
```
