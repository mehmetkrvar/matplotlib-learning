```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years =[2014,2015,2016,2017,2018,2019,2020,2021]
income = [62.5, 67, 68.5, 73, 74.5, 81.5, 68.5, 83.5]
income_ticks=list(range(50,81,2))
```

```python
plt.plot(years,income)
plt.title("John Gelir Grafiği",fontsize=20,color="red",fontname="Times New Roman")
plt.xlabel("Yıllar",fontsize=15,color="blue",fontname="Times New Roman")
plt.ylabel("Gelir (TL)",fontsize=15,color="blue",fontname="Times New Roman")
plt.yticks(income_ticks,[f"{x}k"for x in income_ticks])
plt.grid(True)
plt.show()
'''
plt.title = grafik başlığı
plt.xlabel = x ekseni başlığı
plt.ylabel = y ekseni başlığı
plt.yticks = y ekseni değerleri
plt.grid = grid çizgileri
'''
```
