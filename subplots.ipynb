```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(100)
fig,axs =plt.subplots(2,2,figsize=(10,10))
axs[0,0].plot(x,x)
axs[0,0].set_xlabel('x')
axs[0,0].set_ylabel('y')
axs[0,0].set_title('x-y grafiği')
axs[0,1].plot(x,-x)
axs[1,0].set_xlabel('TEST')
axs[1,0].set_ylabel('y')
axs[1,0].set_title('x-y grafiği')
axs[1,0].plot(x,x**2)
axs[1,1].plot(x,np.log(x))
plt.show()
'''
bu sayede 2x2 lik bir tablo oluşturduk ve 
her birine ayrı ayrı grafik çizdirdik
'''
```
