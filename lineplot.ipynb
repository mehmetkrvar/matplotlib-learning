```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years= [2006 +x for x in range(16)]
print(years)
weights = np.random.randint(50, 100, 16)
print(weights)
print(len(years))
print(len(weights))
```

```python
plt.plot(years, weights,c='g',lw=3,linestyle='--',marker='o',ms=10,mfc='r',mec='b')
'''
years: x ekseni
weights: y ekseni
c: renk 
lw: çizgi kalınlığı
linestyle: çizgi stili
marker: nokta stili
ms: nokta büyüklüğü
mfc: nokta içi renk
mec: nokta kenar rengi
'''
```
